```python
# let's compare the execution speed of this numpy method versus our while loop method
import timeit
```

```python
%timeit np.random.uniform(-90,90,size=1500)
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
%timeit latitudes(1500)
```

```python
#is a for loop faster
def latitudes(size):
    latitudes = []
    x = 0
    for x in range(size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
%timeit latitudes(1500)
```
